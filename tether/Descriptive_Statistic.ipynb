```python
import pandas as pd
import matplotlib.pyplot as plt

tether_data = pd.read_csv('./USDT.csv')

# Calculating count
count = len(tether_data)
print('Count:', count)

# Calculating max
max_price = max(tether_data['Price'])
print('Max:', max_price)

# Calculating min
min_price = min(tether_data['Price'])
print('Min:', min_price)

# Calculating the mean
mean_price = tether_data['Price'].mean()
print('Mean:', mean_price)

# Calculating the median
median_price = tether_data['Price'].median()
print('Median:', median_price)

# Calculating the mode
mode_price = tether_data['Price'].mode()
print('Mode:', mode_price)

# Calculating the standard deviation
std_dev = tether_data['Price'].std()
print('Standard Deviation:', std_dev)

# Calculating the coeficient of deviation
cv = std_dev / mean_price
print('Coefficient of Deviation:', cv)

# Calculating the variance
variance = tether_data['Price'].var()
print('Variance:', variance)

# Calculating skewness
skewness = tether_data['Price'].skew()
print('Skewness:', skewness)

# Calculating kurtosis
kurtosis = tether_data['Price'].kurt()
print('Kurtosis:', kurtosis)

# Calculating the range
price_range = tether_data['Price'].max() - tether_data['Price'].min()
print('Range:', price_range)

# Calculating percentiles (0th, 25th, 50th, 75th, and 100th percentiles)
quantiles = tether_data['Price'].quantile([0, 0.25, 0.5, 0.75, 1])
print('0th Percentile:', quantiles[0])
print('25th Percentile (Q1):', quantiles[0.25])
print('50th Percentile (Median):', quantiles[0.5])
print('75th Percentile (Q3):', quantiles[0.75])
print('100th Percentile:', quantiles[1])
```

```python
plt.figure(figsize=(10, 6))
plt.hist(tether_data['Price'], bins=30, color='skyblue')
plt.title('Tether Price Histogram')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(tether_data['Price'], vert=True)
plt.title('Tether Price Boxplot')
plt.ylabel('Price')
plt.show()
```
