```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Load the dataset
df = pd.read_csv("./USDT.csv")

# Drop the unnecessary columns.
df.drop(['Open', 'High', 'Low', 'Change %'], axis=1, inplace=True)

# Convert the "Date" column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')

# Format the 'Date' column
# df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
                                    
# Sort the DataFrame by date
df = df.sort_values('Date')

# Set Date index
df.set_index('Date', inplace=True)

df
```

```python
# Import necessary libraries
from pydlm import dlm, trend, seasonality, dynamic, autoReg, longSeason

# Create a DLM model
myDLM = dlm(df['Price'])

# Add a trend component
myDLM = myDLM + trend(degree=1, discount=0.9, name='trend')

# Add a seasonality component
myDLM = myDLM + seasonality(period=30, discount=0.9, name='seasonality')

# Fit the model
myDLM.fit()

# Use the model to make predictions
prediction = myDLM.predict(30)

# Print the predictions
print(prediction)
```
